# LlamaTune повторение эксперимента


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# low-dim sensitivity

for n in ('8', '16', '24'):
    results_file = "ycsb_hesbo_{}.csv".format(n)
    results_table = pd.read_csv(results_file)
    plt.plot(results_table.Optimum, label=n)

plt.title('Number of transactions dependent on parameter count')
plt.xlabel('iteration')
plt.ylabel('transactions')
plt.legend()
plt.show()
```

```python
# low-dim sensitivity

for n in ('1', '5', '10', '20'):
    results_file = "ycsb_bucket_{}k.csv".format(n)
    results_table = pd.read_csv(results_file)
    plt.plot(results_table.Optimum, label=n)

plt.title('Number of transactions dependent on bucket count')
plt.xlabel('iteration')
plt.ylabel('transactions')
plt.legend()
plt.show()
```
